```python
def map_timestamps(df, name):
  df = {'start_time': df.iloc[0,0], 'sampling_freq': df.iloc[1,0], 'data': df.iloc[2:,:]}
  df_data = df['data']
  df_data.rename(columns={0: name}, inplace=True)
  sample_rate = df['sampling_freq']
  start = df['start_time']
  stop = start + (len(df['data'])/sample_rate)
  time_series = np.linspace(start, stop, num=len(df['data'])).tolist()
  time_series = [round(ts) for ts in time_series]
  df_data = df['data']
  df_data['time_series'] = time_series
  df_data.set_index('time_series', inplace=True)
  return df_data
```

```python
def read_all_csv(base_path, data_file_name, data_header):
  df_list = []
  for entry in os.listdir(base_path):
    df = pd.read_csv(base_path+entry+'/'+data_file_name+'.csv', header=None)
    df_data = map_timestamps(df, data_header)
    df_list.append(df_data)

  all_data = pd.concat(df_list)
  return all_data
```

```python
all_eda = read_all_csv(base_path, 'EDA', 'ElectroDermalActivity')
all_bvp = read_all_csv(base_path, 'BVP', 'BloodVolumePulse')
all_hr =  read_all_csv(base_path, 'HR', 'HeartRate')
all_temp = read_all_csv(base_path, 'TEMP', 'Temperature')

```

```python
# Import libraries
import glob
import pandas as pd
import re
# Get CSV files list from a folder
capture_path = '/content/drive/MyDrive/Project_Analysis/E4/tobii_capture_data/'
```

```python
def read_all_captures(csv_files):
  df_list = []
  for file in csv_files:
    print(file)
    df = pd.read_csv(file)
    df = df.iloc[np.r_[0:1, -1:0]]
    df['docId'] = 'D'+str(int(file.split('(', 1)[1].split(')')[0])-1)
    df_list.append(df)
  data = pd.concat(df_list)
  data.reset_index(drop=True, inplace=True)
  return data
```

```python
df_list = []
for part in os.listdir(capture_path):
  csv_files = glob.glob(capture_path + part + "/*.csv")
  df = read_all_captures(csv_files)
  df['pId'] = part
  df_list.append(df)

capture_data = pd.concat(df_list)
capture_data.reset_index(drop=True, inplace=True)
```

```python
conditions2 = [
    (capture_data['label'] ==1 ),
    (capture_data['label'] != 1)]

# create a list of the values we want to assign for each condition
class_values2 = [0,1]
categories2 = ['Doc_Start','Doc_End']
# create a new column and use np.select to assign values to it using our lists as arguments
capture_data['read_class'] = np.select(conditions2, class_values2)
capture_data['read_status'] = np.select(conditions2, categories2)
capture_data.drop('label', 1,inplace=True)
```

```python
def get_signal_data(pId, docId, data, start, end, name ):

  l = start+3600
  u = end+3600
  df = pd.DataFrame()
  res = data.loc[l:u]
  df2 = {'participant':pId, 'document':docId, 
         name+'_max': res.iloc[:, 0].max(), name+'_min': res.iloc[:, 0].min(),
         name+'_std': res.iloc[:, 0].std(),name+'_mean': res.iloc[:, 0].mean()}
  df = df.append(df2, ignore_index = True)

  return df

```

```python
from functools import reduce
df_list = []
for i in range(0,len(capture_data),2):
  doc_start = capture_data.loc[i]
  doc_end = capture_data.loc[i+1]
  start_ts = int(doc_start['#timestamp']/1000)
  end_ts = int(doc_end['#timestamp']/1000)
  eda = get_signal_data(doc_start['pId'],doc_start['docId'],all_eda, start_ts,end_ts, 'eda')
  bvp = get_signal_data(doc_start['pId'],doc_start['docId'],all_bvp, start_ts,end_ts, 'bvp')
  hr = get_signal_data(doc_start['pId'],doc_start['docId'],all_hr, start_ts,end_ts, 'hr')
  temp = get_signal_data(doc_start['pId'],doc_start['docId'],all_temp, start_ts,end_ts, 'temp')
  data_frames = [eda, hr, bvp, temp]
  df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['participant','document'],
                                            how='outer'), data_frames)


  df_list.append(df_merged)
data = pd.concat(df_list)
data.reset_index(drop=True,inplace=True)
```
